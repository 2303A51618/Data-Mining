```python
import pandas as pd
```

```python
df= pd.read_csv('/content/employee_data.csv')
print(df)
```

```python

df.isna()
```

```python

df.isnull().sum()
```

```python

df['Age'].mean()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
df['Job_Satisfaction'] = df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mode()[0])

print("\nDataset after handling missing values:")
print(df.head())
```

```python

from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
df[['Age_Scaled', 'Salary_Scaled']] = minmax_scaler.fit_transform(df[['Age', 'Salary']])

print("\nDataset after Min-Max Scaling (sklearn):")
print(df[['Age', 'Age_Scaled', 'Salary', 'Salary_Scaled']].head())
```

```python

df['Job_Satisfaction_Z'] = (df['Job_Satisfaction'] - df['Job_Satisfaction'].mean()) / df['Job_Satisfaction'].std()

df['Work_Hours_Per_Week_Z'] = (df['Work_Hours_Per_Week'] - df['Work_Hours_Per_Week'].mean()) / df['Work_Hours_Per_Week'].std()

print("\nDataset after Z-score Normalization:")
print(df[['Job_Satisfaction', 'Job_Satisfaction_Z', 'Work_Hours_Per_Week', 'Work_Hours_Per_Week_Z']].head())
```

```python
print("\nFinal Preprocessed Dataset:")
print(df.head())
```

```python
df.to_csv('preprocessed_employee_data.csv', index=False)

print("\nPreprocessed dataset saved as 'preprocessed_employee_data.csv'")

```
